```
#change lysanderRDD to dreamRDD
wordsRDD = dreamRDD.map(lambda line: line.split())
wordsRDD.take(5)
```

```
#lysanderRDD to dreamRDD 
wordsRDD = dreamRDD.flatMap(lambda line: line.split())
wordsRDD.take(5)
```

```
wordWithCountRDD = wordsRDD.map(lambda w: (w, 1))
wordWithCountRDD.take(5)
```

```
summedCountRDD = wordWithCountRDD.reduceByKey(lambda x,y: x+y)
summedCountRDD.count()
```

```
summedCountRDD.take(5)
```

```
sortedByKeyRDD = summedCountRDD.sortByKey()
sortedByKeyRDD.take(5)
```

```
a = sortedByKeyRDD.map(lambda (x,y): (y,x))
```
